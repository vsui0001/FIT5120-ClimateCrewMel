#### Getting months data

```python
import requests # importing libraries 
from bs4 import BeautifulSoup
import urllib.request
import time
```

```python
total_list = []
URL1 = 'https://gardenate.com/?month='
for x in range(1,13):
    URL = URL1+str(x)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    temp = soup.find('div', {"class" : 'plant-list'})
    total = temp.findAll('a')
    plant_list  = []
    for x in total:
        t = x['href'].split('/')
        if len(t) == 3:
            if t[1] == 'plant':
                t1 = t[2].replace("%2B", " ")
                t1 = t1.replace("%2C", ",")
                plant_list.append(t1)
    total_list.append(plant_list) # getting months data and apending it to out main data frame 
```
